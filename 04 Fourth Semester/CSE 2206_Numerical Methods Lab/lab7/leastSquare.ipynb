```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([2.5, 2.9, 4.6, 5.5,6.1, 7.9])

m = len(x)

sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)

a = (m * sum_xy - sum_x * sum_y) / (m * sum_x2 - sum_x ** 2)
b = (sum_y - a * sum_x) / m

print(f"Coefficients: a = {a}, b = {b}")

plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, a * x + b, color='red', label='Fitted line')
plt.grid()
plt.title('Least Squares Fit')
plt.legend()
plt.show()

```
